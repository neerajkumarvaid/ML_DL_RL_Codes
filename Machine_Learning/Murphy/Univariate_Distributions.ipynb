Univariate Gaussian Distribution

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np


rv = norm(0,1)

fig, ax = plt.subplots(1,2,figsize =(10,5))

X = np.linspace(-3,3, 500)
ax[0].plot(X, rv.pdf(X))
ax[0].set_title("Gaussian PDF")
ax[1].plot(X, rv.cdf(X))
ax[1].set_title("Gaussian CDF")
plt.show()
```

How to draw samples from a normal distribution?

```python
np.random.seed(30) # seed to ensure the repeatability of the following example

mu = 1.1 # specify the mean of the distribution
sigma = 0.1 # specify the variance of the distribution
dist = norm(loc = mu, scale = sigma) # create a fixed normal distribution

N = 10 # number of points to be sampled
x = dist.rvs(size = N) # draw N samples
print(x.shape)
print(x)

np.random.seed(30)
x2 = norm(loc = mu, scale = sigma).rvs(size = N)

assert np.allclose(x,x2)
```

```python
# pdf, cdf, and inverse cdf

logprob = dist.logpdf(x)
print(logprob)
```

```python
p = dist.cdf(x)
x3 = dist.ppf(p) # inverse cdf
```

```python
assert np.allclose(x,x3)
```

Gamma Distribution

```python
from scipy.stats import gamma

x = np.linspace(0,7,100)
b = 1
plt.figure()
for a in [1,1.5,2]:
  y = gamma.pdf(x, a, scale = 1/b, loc = 0)
  plt.plot(x,y)
plt.legend(['a = %.1f, b = 1' % a for a in [1,1.5,2]])
plt.title('Gamma (a,b) Distributions')
plt.show()
```
